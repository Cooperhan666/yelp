```python
sp = spark.createDataFrame(df)
```

```python
sp.printSchema()
```

```python
restaurant_style_indexer = StringIndexer(inputCol='restaurant_style',outputCol='restaurant_style_index')
restaurant_style_encoder = OneHotEncoder(inputCol='restaurant_style_index',outputCol='restaurant_styleVec')
restaurant_WiFi_indexer = StringIndexer(inputCol='restaurant_WiFi',outputCol='restaurant_WiFi_index')
restaurant_WiFi_encoder = OneHotEncoder(inputCol='restaurant_WiFi_index',outputCol='restaurant_WiFiVec')
garage_parking_indexer = StringIndexer(inputCol='garage_parking',outputCol='garage_parking_index')
garage_parking_encoder = OneHotEncoder(inputCol='garage_parking_index',outputCol='garage_parkingVec')
street_parking_indexer = StringIndexer(inputCol='street_parking',outputCol='street_parking_index')
street_parking_encoder = OneHotEncoder(inputCol='street_parking_index',outputCol='street_parkingVec')
lot_parking_indexer = StringIndexer(inputCol='lot_parking',outputCol='lot_parking_index')
lot_parking_encoder = OneHotEncoder(inputCol='lot_parking_index',outputCol='lot_parkingVec')
valet_parking_indexer = StringIndexer(inputCol='valet_parking',outputCol='valet_parking_index')
valet_parking_encoder = OneHotEncoder(inputCol='valet_parking_index',outputCol='valet_parkingVec')
restaurant_noise_level_indexer = StringIndexer(inputCol='restaurant_noise_level',outputCol='restaurant_noise_level_index')
restaurant_noise_level_encoder = OneHotEncoder(inputCol='restaurant_noise_level_index',outputCol='restaurant_noise_levelVec')
assembler = VectorAssembler(inputCols=['restaurant_styleVec','restaurant_stars','restaurant_WiFiVec','garage_parkingVec','street_parkingVec','lot_parkingVec','restaurant_noise_levelVec'], outputCol="classificationFeatures")
```

```python
pipeline = Pipeline(stages=[restaurant_style_indexer, restaurant_style_encoder, restaurant_WiFi_indexer, restaurant_WiFi_encoder, garage_parking_indexer, garage_parking_encoder, street_parking_indexer, street_parking_encoder, lot_parking_indexer, lot_parking_encoder, restaurant_noise_level_indexer, restaurant_noise_level_encoder, assembler])
pipeline_model = pipeline.fit(sp)
pipe_sp = pipeline_model.transform(sp)
pipe_sp = pipe_sp.select('user_stars', 'classificationFeatures')
train_data, test_data = pipe_sp.randomSplit([0.7,0.3])
```

```python
mlr_model = LogisticRegression(featuresCol='classificationFeatures',labelCol='user_stars', family='multinomial',maxIter=10, regParam=0, elasticNetParam=0)
mlr_model = mlr_model.fit(train_data)
results = mlr_model.transform(test_data)
results.select("prediction", "user_stars", "classificationFeatures").show()
evaluator = MulticlassClassificationEvaluator(
    labelCol="user_stars", predictionCol="prediction", metricName="accuracy")
lRAccuracy = evaluator.evaluate(results)
print("Accuray = %g " %lRAccuracy)
print("Test Error = %g " %(1.0 - lRAccuracy))
```
